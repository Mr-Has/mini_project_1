```python
def lr(x, y, m_new, c_new, learning_rate, epoch):
    N = float(len(y))
    for i in range(epoch):
        y_pred = (m_new * x) + c_new
        cost = sum([t ** 2 for t in (y - y_pred)])/N
        c_grad = - (2 / N) * sum (y - y_pred)
        m_grad = - (2 / N) * sum (x * (y - y_pred))
        m_new = m_new - (learning_rate * m_grad)
        c_new = c_new - (learning_rate * c_grad)
        line = m_new * x + c_new
        plt.scatter(x, y, c='r')
        plt.plot(x, line)
        plt.grid()
        plt.pause(3)
        print(cost)
    return c_new,m_new,cost
    
```

```python
def calculate_cost(theta, X_train, Y_train):

    m = len(Y_train)
    predictions = X_train.dot(theta)
    cost = (1/2 * m) * np.sum(np.square(predictions - Y_train))
    return cost
    
```

```python
def gradient_descent(X_train ,Y_train ,theta_history ,learning_rate=0.01 ,iterations=100):
    m = len(Y_train)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations,2))
    for it in range(iterations):
        
        prediction = np.dot(X_train, theta)
        
        theta = theta - (1 / m) * learning_rate * (X_train.T.dot((prediction - Y_train)))
        theta_history[it,:] = theta.T
        cost_history[it]  = cal_cost(theta,X_train, Y_train)
        plt.scatter(X_train, Y_train, c='r')
        plt.plot(X_train, cost_history)
        plt.grid()
        plt.pause(1)
    return theta, cost_history, theta_history
 
    
```
